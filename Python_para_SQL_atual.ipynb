```python
'''
Esta consulta acontece no nível instância, e serve para ver quais bancos
temos em nossa instância
'''

# Estabelecimento da conexão com o banco de dados usando as informações fornecidas
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
)
cursor = connection.cursor() # Criação de um cursor para executar consultas SQL no banco de dados
query = 'show databases;' # Definição da consulta SQL a ser executada
cursor.execute(query) # Execução da consulta no banco de dados
result = cursor.fetchall() # Recuperação de todos os resultados da consulta
tables = pd.DataFrame(result, columns=cursor.column_names) # Criação de um DataFrame Pandas usando os resultados da consulta e os nomes das colunas do cursor
display(tables) # Exibição do DataFrame Pandas contendo os resultados da consulta
cursor.close() # Fechamento do cursor após a conclusão da consulta
connection.close() # Fechamento da conexão após a consulta
```

# ** DEFINIÇÃO DAS INFORMAÇÕES DE CONEXÃO AO BANCO DE DADOS**
## ** CRIAÇÃO DO BANCO DE DADOS E DOS ATRIBUTOS**

```python
# Definição das informações de conexão ao banco de dados
host = '34.42.189.175'
user = 'root'
password = ''
database = 'sextou'
```

```python
'''
Objetivo: Criar uma tabela no banco de dados
'''
# Estabelecimento da conexão com o banco de dados usando as informações fornecidas
connection = mysql.connector.connect(
    host=host,
    database=database,
    user=user,
    password=password
    )
cursor = connection.cursor() # Criação de um cursor para executar consultas SQL no banco de dados
query = '''
create table Cliente (
  id_cliente INT PRIMARY KEY,
  nome VARCHAR (50),
  endereco VARCHAR (100)
);
''' # Definição da consulta SQL a ser executada
cursor.execute(query) # Execução da consulta no banco de dados
cursor.close() # Fechamento do cursor após a conclusão da consulta
connection.close() # Fechamento da conexão após a consulta
```

```python
'''
Objetivo: Criar uma tabela no banco de dados
'''
# Estabelecimento da conexão com o banco de dados usando as informações fornecidas
connection = mysql.connector.connect(
    host=host,
    database=database,
    user=user,
    password=password
    )
cursor = connection.cursor() # Criação de um cursor para executar consultas SQL no banco de dados
query = '''
CREATE TABLE Pedido (
    id_pedido INT PRIMARY KEY,
    id_cliente INT,
    descricao VARCHAR(100),
    FOREIGN KEY (id_cliente) REFERENCES Cliente(id_cliente)
);
''' # Definição da consulta SQL a ser executada
cursor.execute(query) # Execução da consulta no banco de dados
cursor.close() # Fechamento do cursor após a conclusão da consulta
connection.close() # Fechamento da conexão após a consulta
```

```python
'''
Objetivo: Criar uma tabela no banco de dados
'''
# Estabelecimento da conexão com o banco de dados usando as informações fornecidas
connection = mysql.connector.connect(
    host=host,
    database=database,
    user=user,
    password=password
    )
cursor = connection.cursor() # Criação de um cursor para executar consultas SQL no banco de dados
query = '''
CREATE TABLE Produto (
    id_produto INT PRIMARY KEY,
    nome VARCHAR(50),
    preco DECIMAL(10, 2)
);
''' # Definição da consulta SQL a ser executada
cursor.execute(query) # Execução da consulta no banco de dados
cursor.close() # Fechamento do cursor após a conclusão da consulta
connection.close() # Fechamento da conexão após a consulta
```

```python
'''
Objetivo: Criar uma tabela no banco de dados
'''
# Estabelecimento da conexão com o banco de dados usando as informações fornecidas
connection = mysql.connector.connect(
    host=host,
    database=database,
    user=user,
    password=password
    )
cursor = connection.cursor() # Criação de um cursor para executar consultas SQL no banco de dados
query = '''
CREATE TABLE ItemPedido (
    id_item INT PRIMARY KEY,
    id_pedido INT,
    id_produto INT,
    quantidade INT,
    FOREIGN KEY (id_pedido) REFERENCES Pedido(id_pedido),
    FOREIGN KEY (id_produto) REFERENCES Produto(id_produto)
);
''' # Definição da consulta SQL a ser executada
cursor.execute(query) # Execução da consulta no banco de dados
cursor.close() # Fechamento do cursor após a conclusão da consulta
connection.close() # Fechamento da conexão após a consulta
```

```python
'''
Objetivo: Criar uma tabela no banco de dados
'''
# Estabelecimento da conexão com o banco de dados usando as informações fornecidas
connection = mysql.connector.connect(
    host=host,
    database=database,
    user=user,
    password=password
    )
cursor = connection.cursor() # Criação de um cursor para executar consultas SQL no banco de dados
query = '''
CREATE TABLE Categoria (
    id_categoria INT PRIMARY KEY,
    nome VARCHAR(50)
);
''' # Definição da consulta SQL a ser executada
cursor.execute(query) # Execução da consulta no banco de dados
cursor.close() # Fechamento do cursor após a conclusão da consulta
connection.close() # Fechamento da conexão após a consulta
```

```python
'''
Objetivo: Criar uma tabela no banco de dados
'''
# Estabelecimento da conexão com o banco de dados usando as informações fornecidas
connection = mysql.connector.connect(
    host=host,
    database=database,
    user=user,
    password=password
    )
cursor = connection.cursor() # Criação de um cursor para executar consultas SQL no banco de dados
query = '''
CREATE TABLE ProdutoCategoria (
    id_produto INT,
    id_categoria INT,
    PRIMARY KEY (id_produto, id_categoria),
    FOREIGN KEY (id_produto) REFERENCES Produto(id_produto),
    FOREIGN KEY (id_categoria) REFERENCES Categoria(id_categoria)
);
''' # Definição da consulta SQL a ser executada
cursor.execute(query) # Execução da consulta no banco de dados
cursor.close() # Fechamento do cursor após a conclusão da consulta
connection.close() # Fechamento da conexão após a consulta
```

# ** CALCULAR A MÉDIA

```python
'''
imagine que voce deseja fazer uma consulta que retorne a quantidade média
do atributo quantidade da tabela ItemPedido, arredondando amédia para 2 casas
decimais
'''
# Abertura da conexão para uma nova consulta
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
)
cursor = connection.cursor()  # Criação de um cursor para executar consultas SQL no banco de dados
# Definição da consulta SQL a ser executada
query = '''
SELECT AVG(quantidade) AS qtde_media,
round(AVG(quantidade),2) AS qtde_media_arredondada
FROM ItemPedido;
'''
cursor.execute(query) # Execução da consulta no banco de dados
result = cursor.fetchall() # Recuperação de todos os resultados da consulta
media = pd.DataFrame(result, columns=cursor.column_names) # Criação de um DataFrame Pandas usando os resultados da consulta e os nomes das colunas do cursor
display(media) # Exibição do DataFrame Pandas contendo os resultados da consulta
cursor.close() # Fechamento do cursor após a conclusão da consulta
connection.close() # Fechamento da conexão após a consulta
'''
Resumo do comando: Seleciona as colunas: quantidade da tabela ItemPedido,
retorna o valor arredondado (ROUND) médio da coluna (AVG), e armazena os
resultados da consulta em um dataframe Pandas chamado media
'''
```

```python
'''
Agora Imagine que você deseja fazer uma consulta na tabela Cliente que retorne
a frequência de cada endereço
'''
# Abertura da conexão para uma nova consulta
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
)
cursor = connection.cursor()  # Criação de um cursor para executar consultas SQL no banco de dados
# Definição da consulta SQL a ser executada
query = '''
SELECT endereco
     , count(*) as freq
  FROM Cliente
 GROUP BY endereco;
'''
cursor.execute(query) # Execução da consulta no banco de dados
result = cursor.fetchall() # Recuperação de todos os resultados da consulta
grupo = pd.DataFrame(result, columns=cursor.column_names) # Criação de um DataFrame Pandas usando os resultados da consulta e os nomes das colunas do cursor
display(grupo) # Exibição do DataFrame Pandas contendo os resultados da consulta
cursor.close() # Fechamento do cursor após a conclusão da consulta
connection.close() # Fechamento da conexão após a consulta
'''
Resumo do comando: Comando que seleciona o atributo endereço, cria e seleciona a
coluna count renomeada com freq da tabela Cliente e agrupa por endereco,
e armazena os resultados da consulta em um dataframe Pandas chamado grupo
'''
```

```python
'''
Agora Imagine que você deseja fazer uma consulta na tabela Cliente que retorne
a frequência de cada endereço, mas apenas para aqueles endereços que tiverem
frequência maior do 1
'''
# Abertura da conexão para uma nova consulta
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
)
cursor = connection.cursor()  # Criação de um cursor para executar consultas SQL no banco de dados
# Definição da consulta SQL a ser executada
query = '''
SELECT endereco
     , count(*)
  FROM Cliente
 GROUP BY endereco
 HAVING count(*) > 1;
'''
cursor.execute(query) # Execução da consulta no banco de dados
result = cursor.fetchall() # Recuperação de todos os resultados da consulta
tendo = pd.DataFrame(result, columns=cursor.column_names) # Criação de um DataFrame Pandas usando os resultados da consulta e os nomes das colunas do cursor
display(tendo) # Exibição do DataFrame Pandas contendo os resultados da consulta
cursor.close() # Fechamento do cursor após a conclusão da consulta
connection.close() # Fechamento da conexão após a consulta
'''
Resumo do comando: Consulta que seleciona endereco e contangem de endereco da
tabela Cliente, Agrupado por endereço tendo apenas (HAVING) os edereços que se
repetem pelo menos duas vezes. e armazena os resultados da consulta
em um dataframe Pandas chamado tendo
'''
```
