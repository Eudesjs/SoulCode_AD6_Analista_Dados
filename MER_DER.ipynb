```python
'''
Agora que já temos o modelo entidade-relacionamento e o diagrama entidade relacionamento
vamos transformá-los em um banco de dados chamado faculdade

1 - criar o banco de dados faculdade pelo shell
create database faculdade;

2 - apontar para o banco criado
use faculdade;

3 - criar as tabelas

create table alunos (
  id_aluno INT PRIMARY KEY,
  nome VARCHAR (50),
  curso VARCHAR (100),
  semestre INT
);

após criar a tabela podemos conferir se ela foi criada pelo comando:
show tables;

além disso, se desejarmos saber o schema da tabela criada basta usar:
desc nome_da_tabela;


create table professores (
  id_professor INT PRIMARY KEY,
  nome VARCHAR (50),
  curso VARCHAR (100),
  disciplina VARCHAR (100)
);

create table cursos (
  id_curso INT PRIMARY KEY,
  id_aluno INT,
  curso VARCHAR (100),
  FOREIGN KEY (id_aluno) REFERENCES alunos(id_aluno)
);

create table disciplinas (
  id_disciplina INT PRIMARY KEY,
  id_professor INT,
  id_curso INT,
  disciplina VARCHAR (100),
  FOREIGN KEY (id_professor) REFERENCES professores(id_professor),
  FOREIGN KEY (id_curso) REFERENCES cursos(id_curso)
);

'''
```

```python
'''
    TABLE_NAME AS 'Tabela',
    COLUMN_NAME AS 'Coluna',
    CONSTRAINT_NAME AS 'Nome da Restrição',
    REFERENCED_TABLE_NAME AS 'Tabela Referenciada',
    REFERENCED_COLUMN_NAME AS 'Coluna Referenciada'
    FROM
    INFORMATION_SCHEMA.KEY_COLUMN_USAGE
    WHERE
    TABLE_SCHEMA = 'faculdade'
     AND REFERENCED_TABLE_NAME IS NOT NULL;
'''
```
