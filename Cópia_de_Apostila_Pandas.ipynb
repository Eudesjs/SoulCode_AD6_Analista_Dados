```python
# renomear colunas método 01
nome = ['Douglas','Daniela','Pedro','Maria','Eduardo','Ester']
idade = [45,7,65,64,42,37]
altura = [1.85,1.23,1.75,1.67,1.82,1.73]
peso = [70,22,87,64,96,68]
sexo = ['M','F','M','F','M','F']
df = pd.DataFrame(zip(nome,idade,altura,peso,sexo),
                  columns=['nome','idade','altura','peso','sexo'])
'''
Vamos renomear as colunas peso (massa) e sexo (genero)
'''
df.rename(columns={'peso':'massa','sexo':'genero'},inplace=True)
display(df)
```

```python
'''
No bloco anterior, ao aplicar o groupby nós geramos um objeto que possui uma única
dimensão, mesmo mostrando " duas colunas", assim, se nosso objetivo for guardar
essas frequências em uma tabela, devemos fazer o seguinte:
'''
import pandas as pd
nome = ['Douglas','Daniela','Pedro','Maria','Eduardo','Ester','Zé']
idade = [45,7,65,54,25,37,16]
altura = [1.85,1.23,1.75,1.67,1.82,1.73,1.68]
peso = [70,22,87,64,96,68,70]
sexo = ['M','F','M','F','M','F','M']
df = pd.DataFrame(zip(nome,idade,altura,peso,sexo),
                  columns=['nome','idade','altura','peso','sexo'])
df['imc'] = round(df['peso']/df['altura']**2,2)
df.rename(columns={'peso':'massa','sexo':'genero'},inplace=True)
faixas = ['01 - 10','11 - 20','21 - 30','31 - 40','41 - 50','51 - 60','61 - 70']
df['faixa_etaria'] = pd.cut(x=df['idade'],bins=[0,10,20,30,40,50,60,70],labels=faixas)
'''
1 - gerar uma lista com os valores únicos da variável que estamos contando e
armazená-la em uma variável
'''
xpto = df['faixa_etaria'].unique()

'''
2 - Realizar o groupby e armazená-lo em uma variável
'''
freq = df.groupby('faixa_etaria')['faixa_etaria'].count()

'''
3 - Criar um dataframe com as duas variáveis armazenadas anteriormente
'''
df3 = pd.DataFrame(zip(xpto,freq),columns=['Faixa Etária','Frequência'])
display(df3)
'''
Nesse processo temo um PROBLEMA:
ao gerar um objeto pelo .unique() teremos uma lista só com os valores únicos
que aparecem na variável faixa etária.
No entanto quando geramos o segundo objeto pelo .groupby podemos ter uma lista
maior que a lista gerada pelo unique, pois o .groupby contará também as categorias
listadas cuja frequencia for 0 , ou seja, aqueles que não aparecerão na lista do
.unique, e como resultado teremos duas listas que terão tamanhos diferentes
e ao usarmos o .pd.DataFrame a união será feita com base na lista de menos elementos
'''
df3 = df3.sort_values(by='Faixa Etária')
display(df3)
```

```python
'''
Vamos criar a frequência percentual para a variável faixa etária pelo método
value_counts
'''
import pandas as pd
nome = ['Douglas','Daniela','Pedro','Maria','Eduardo','Ester']
idade = [45,7,65,64,42,37]
altura = [1.85,1.23,1.75,1.67,1.82,1.73]
peso = [70,22,87,64,96,68]
sexo = ['M','F','M','F','M','F']
df = pd.DataFrame(zip(nome,idade,altura,peso,sexo),
                  columns=['nome','idade','altura','peso','sexo'])
df['imc'] = round(df['peso']/df['altura']**2,2)
df.rename(columns={'peso':'massa','sexo':'genero'},inplace=True)
faixas = ['01 - 10','11 - 20','21 - 30','31 - 40','41 - 50','51 - 60','61 - 70']
df['faixa_etaria'] = pd.cut(x=df['idade'],bins=[0,10,20,30,40,50,60,70],labels=faixas)
freq = df['faixa_etaria'].value_counts() # aqui estamos criando um objeto com a contagem de todos os valores únicos da variável de interesse
percentual = df['faixa_etaria'].value_counts(normalize=True)*100 # aqui estamos criando um objeto com a contagem percentual de todos os valores únicos da variável de interesse
idades = pd.DataFrame({'Frequência Absoluta':freq,'Frequência Percentual':percentual}) # aqui estamos criando um dataframe com os dois objetos criados anteiormente
idades = idades.sort_values(by='faixa_etaria')
idades.reset_index(inplace=True)

'''
Quando temos uma variavél quantitativa a distribuição de frequência deve ter
duas colunas a mais:
1 - Frequencia acumulada
2 - Frequencia acumulada %
'''
idades['Frequência Acumulada'] = idades['Frequência Absoluta'].cumsum() # Esta linha cria a coluna de frequencia acumulada
idades['Frequência Acumulada %'] = (idades['Frequência Acumulada']/idades['Frequência Absoluta'].sum())*100

idades.loc[7] = ['TOTAL',idades['Frequência Absoluta'].sum(),
                  idades['Frequência Percentual'].sum(),'-','-'] # Aqui estamos inserindo uma linha de total na tabela de frequências
display(idades)

```

```python
import pandas as pd
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vTFGertCc67cL9ojU5rRq8qyCBojGKrH7m2zU7hqYJ1AKRm-oBpTNjRhGztmhYrzJApq3-FNbx0EPz3/pub?gid=0&single=true&output=csv')
faixas = ['10001 - 12000', '12001 - 14000', '14001 - 16000', '16001 - 18000', '18001 - 20000']
df['faturamentos'] = pd.cut(x=df['faturamento'],bins=[10000,12000,14000,16000,18000,20000],labels=faixas)
freq = df['faturamentos'].value_counts() # aqui estamos criando um objeto com a contagem de todos os valores únicos da variável de interesse
percentual = df['faturamentos'].value_counts(normalize=True)*100 # aqui estamos criando um objeto com a contagem percentual de todos os valores únicos da variável de interesse
fatur = pd.DataFrame({'Frequência Absoluta':freq,'Frequência Percentual':percentual}) # aqui estamos criando um dataframe com os dois objetos criados anteiormente
fatur = fatur.sort_values(by='faturamentos')
fatur.reset_index(inplace=True)

'''
Quando temos uma variavél quantitativa a distribuição de frequência deve ter
duas colunas a mais:
1 - Frequencia acumulada
2 - Frequencia acumulada %
'''
fatur['Frequência Acumulada'] = fatur['Frequência Absoluta'].cumsum() # Esta linha cria a coluna de frequencia acumulada
fatur['Frequência Acumulada %'] = (fatur['Frequência Acumulada']/fatur['Frequência Absoluta'].sum())*100 # Esta linha cria a coluna de frequencia acumulada %

fatur.loc[5] = ['TOTAL',fatur['Frequência Absoluta'].sum(),
                  fatur['Frequência Percentual'].sum(),'-','-'] # Aqui estamos inserindo uma linha de total na tabela de frequências
display(fatur)
```
