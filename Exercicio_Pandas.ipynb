```python
import pandas as pd
```

```python
nome = ['Douglas','Daniela','Pedro','Maria','Eduardo','Ester']
idade = [45,7,65,64,42,37]
altura = [1.85,1.23,1.75,1.67,1.82,1.73]
peso = [70,22,87,64,96,68]
```

```python
'''
Tarefa para 23/04/2024:

Criar as medidas:
1 - AIQ
2 - LI
3 - LS

e inserí-las nessa ordem no df medidas
'''
# CRIAR UMA DATAFRAME USE O METODO BIBLIOTECA
df = pd.DataFrame({
    'Nome': nome,
    'Idade': idade,
    'Altura': altura,
    'Peso': peso
})

# CALCULAR AIQ (ALTURA DIVIDIDO PELO PESO)
df['AIQ'] = df['Altura'] / df['Peso']

# CALCULAR LI (IDADE MENOS 2 DESVIOS PADRÃO)
mean_age = df['Idade'].mean()
std_age = df['Idade'].std()
df['LI'] = mean_age - 2 * std_age

# CALCULAR LS (IDADE MAIS 2 DESVIOS PADRÃO)
df['LS'] = mean_age + 2 * std_age

# EXIBIR O DATAFRAME COM AS MEDIDAS
display(df)
```
